```python
training_Data=data[['PetalLengthCm','SepalWidthCm','PetalWidthCm','SepalLengthCm']]
testing_Data=data['Species']
x_train,x_test,y_train,y_test=train_test_split(training_Data,testing_Data,test_size=0.2,random_state=42)
```

```python
tree=DecisionTreeClassifier()
tree.fit(x_train,y_train)
```

```python
tree.score(x_train,y_train)
```

```python
xgb_clf = RandomForestClassifier()
xgb_clf.fit(x_train, y_train)
```

```python
xgb_clf.score(x_train, y_train)
```

```python
LG = LogisticRegression()
LG.fit(x_train, y_train)
```

```python
LG.score(x_train, y_train)
```

```python
predictions=LG.predict(x_test)
```

```python

print("Confusion Matrix:")
print(confusion_matrix(y_test, predictions))

print("Classification Report")
print(classification_report(y_test, predictions))
```
